```python
from UTILITY_quickstart import *
```

```python
importedDefaultSettings = loadConfig("setLattice_configs/2024-10-22_oneBunch_baseline3.yml")
```

```python
csrTF = True



inputBeamFilePathSuffix = importedDefaultSettings["inputBeamFilePathSuffix"]

tao = initializeTao(
    inputBeamFilePathSuffix = inputBeamFilePathSuffix,
    
    csrTF = csrTF,
    #numMacroParticles=1e4,
    scratchPath = "/tmp",
    randomizeFileNames = True
)

setLattice(tao, **importedDefaultSettings)



```

```python
trackBeam(tao, trackEnd = "PR10571", **importedDefaultSettings)

POff = getBeamAtElement(tao, "PR10571")

display(plotMod(POff, 'x', 'xp',  bins=300, xlim =(-2e-3, 2e-3), ylim = (-1e-3,1e-3)))
display(plotMod(POff, 'z', 'pz', bins=300, xlim =(-1.5e-3, 1.5e-3), ylim = (122e6,124e6) ))
plt.clf()
```

```python
tao.cmd("set ele * space_charge_method = slice")

trackBeam(tao, trackEnd = "PR10571", **importedDefaultSettings)
PSlice = getBeamAtElement(tao, "PR10571")

display(plotMod(PSlice, 'x', 'xp',  bins=300, xlim =(-2e-3, 2e-3), ylim = (-1e-3,1e-3)))
display(plotMod(PSlice, 'z', 'pz', bins=300, xlim =(-1.5e-3, 1.5e-3), ylim = (122e6,124e6) ))
plt.clf()
```

```python
tao.cmd("set ele * space_charge_method = fft_3d")

trackBeam(tao, trackEnd = "PR10571", **importedDefaultSettings)
PFFT = getBeamAtElement(tao, "PR10571")

display(plotMod(PFFT, 'x', 'xp',  bins=300, xlim =(-2e-3, 2e-3), ylim = (-1e-3,1e-3)))
display(plotMod(PFFT, 'z', 'pz', bins=300, xlim =(-1.5e-3, 1.5e-3), ylim = (122e6,124e6) ))
plt.clf()
```

```python
POffSpecs = getBeamSpecs(POff)
PSliceSpecs = getBeamSpecs(PSlice)
PFFTSpecs = getBeamSpecs(PFFT)
```
